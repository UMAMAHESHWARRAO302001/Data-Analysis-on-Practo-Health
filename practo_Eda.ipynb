```python
import pandas as pd
```

```python
df=pd.read_csv("practo.csv")
```

```python
df.head()
```

```python
df['Specialist'] = df['Specialist'].replace('Gastroenterologist.*', 'Gastroenterologist', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('Dentist.*', 'Dentist', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('Gynecologist/Obstetrician.*', 'Gynecologist/Obstetrician', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('General Physician.*', 'General Physician', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('Ear-Nose-Throat \(ENT\) Specialist.*', 'Ear-Nose-Throat(ENT)', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('Dermatologist.*', 'Dermatologist', regex=True)
```

```python
df['Specialist'] = df['Specialist'].replace('Cardiologist.*', 'Cardiologist', regex=True)
```

```python
df['Specialist'].value_counts()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Specialists'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Experience'] = df['Experience'].str.replace(r'(\d+\s*years)\s*experience\s*overall', r'\1')
```

```python
df.drop(['Specialists'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.to_csv('practoHealtheda.csv',index=False)
```
